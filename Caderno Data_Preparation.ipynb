## Select Data

Select Data é a task onde você efetivamente irá selecionar os dados que são relevantes para o teu Dataset.

```python
# Excluindo colunas irrelavantes e selecionando as importantes
df = df.drop(columns = ['Unnamed: 0','name','category','goal','pledged','usd pledged'])
```

```python
df.head()
```

## Clean Data

Clean Data é a task onde você vai se aprofundar na qualidade dos dados, iniciar a limpeza dos dados podendo selecionar subconjuntos, selecionar alguns padrões e inclusive fazer estimativas.

OUTPUT: Descrever os os métodos e decisões tomadas para resolver os problemas de qualidade encontrados na task de Verify Data Quality na fase de Data Understanding.

```python
# Se a função for execultada sem a seleção da coluna desejada, ela retorna o resultado que será adicinado no DF.
# df['usd_goal_real'].apply(lambda x: x.strip('USD '))
```

```python
# Eliminando palavras
df['usd_goal_real'] = df['usd_goal_real'].apply(lambda x: x.strip('USD '))
```

```python
# RegEx
# Regextester.com
# Exemplo de CPF em regEx
# 999.999.999-99 == '\d\d\d.\d\d\d.\d\d\d-\d\d'
```

```python
# Eliminado a hora, minutos e segundos da data com regEx
df['launched'] = df['launched'].str.replace(' \d\d:\d\d', '', regex=True)
```

## Format Data

Format Data é a formatação de dados que não mudam seu significado porém facilitam o entendimento e a modelagem.

OUTPUT: Dataset reformatado com dados que fazem sentido. ( O identificador único do Dataset deve ser a primeira coluna)

```python
df.dtypes
```

```python
# Tranformando uma string em números inteiros
df['usd_goal_real'] = df['usd_goal_real'].astype('int64')
```

```python
# Datetime format codes
# Transformando a deadline e launched de string para data
df['deadline'] = pd.to_datetime(df['deadline'], format='%d/%m/%y')
df['launched'] = pd.to_datetime(df['launched'], format='%d/%m/%y')
```

```python
df.dtypes
```

## Construct Data

Construct Data é a task onde você vai utilizar de técnicas para construir novas features. As novas features podem ser geradas da combinação de outras features ou registros inteiros novos ou valores transformados.

OUTPUT: Exemplificar e descreva motivos de quais foram as novas features criadas e como foram geradas. EX: área= comprimento*largura (Foi criada a feature area, e ela foi criada através da multiplicação das features comprimento e largura.)

```python
# Calculando o tempo em dias que produto ficou venda, e criando uma nova variável 
df['time_range'] = (df['deadline'] - df['launched']).dt.days
```

```python
df3 = pd.read_csv('invested.csv')
```

```python
df3.head()
```

```python
# Alguns ID apresentam mais de 1 investidor
df3.sort_values(by= ['ID'])
```

```python
# One hot enconding
# Tranformar uma string em uma variável boleana (0 ou 1)
df3 = pd.get_dummies(df3, columns= ['backedLocation'])
```

```python
df3.head()
```

```python
# Fazendo a agregação com os valores formatados em 0 e 1
df3 = df3.groupby(['ID']).agg({'age':'mean','backedLocation_BR':'sum','backedLocation_GBK':'sum','backedLocation_US':'sum'})
```

```python
df3.head()
```

```python
df = df.merge(df3, how='right',on=['ID'])
```
