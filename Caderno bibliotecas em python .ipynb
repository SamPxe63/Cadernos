# Arrays em Numpy

Entendendo na prática o que é uma array:

As dimensões são chamadas de axes.

O número de axes é chamado de rank.

## Criando um Array

Existem três formas de criar um array em numpy. São eles:

np.arange()

np.array()

np.linspace()

```python
!pip install numpy
import numpy as np
```

```python
a = np.array([1,2,3])
print(a)
print()

b = np.arange(1,10+1)
print(b)
print()

c = np.linspace(1,10, num=5)
print(c)
print()

d = np.ones([2,2]) # Cria um array somente com números 1
print(d)
print()

e = np.zeros_like([2,2]) # Cria um array somento com números 0
print(e)
```

```python
# Retorna o rank (número de axes, ou seja, a quantidade de dimensões)
a.ndim
```

```python
# Quantidade de elementos presentes na matriz
a.size
```

## Alterando o formato de um Array

.reshape: Deve ser setado a quantidade de linhas e colunas requeridas. Altera a quantidade de dimensões. (Ex: 10,10--> 10 linhas com 10 colunas: 2 Dimensões)

.[np.newaxis]: Não necessita de valor setado. Altera somente as colunas. (Ex: 1,100--> 1 linha com 100 colunas: 2 Dimensões)

```python
# reshape
b = a.reshape(10,10)
print(b)
```

```python
# Valor de linhas e colunas
b.shape
```

```python
# Número de dimensões
b.ndim
```

```python
# newaxis
c = a[np.newaxis]
print(c)
```

```python
# Valor de linhas e colunas
c.shape
```

```python
# Número de dimensões
c.ndim
```

## Funções em Arrays

### Iterando elementos

```python
import numpy as np
d = np.array([5,7,8,6,1,10,16,18,50,24,25,89,100,63,41,22,28,33,35,0])
print(d)

# Iterando os elementos da array
for element in d.flat:
  print(element)
```

### Ordenando elementos

```python
# Ordenando os elementos da array

d2 = np.sort(d)
print(d2)
```

### Juntando arrays

```python
# Juntando arrays

e = np.array([2,3,4,5,6])
f = np.array([7,8,9,10,11])

ef = np.concatenate([e,f])

print(ef)
ef.shape # Consultando as dimensões
```

```python
# Juntando arrays horizontalmente

ef2 = np.hstack((e,f))
print(ef2)
ef2.shape # Consultando as dimensões
# Juntando arrays verticalmente

ef3 = np.vstack((e,f))
print(ef3)
ef3.shape # Consultando as dimensões
```

### Média em arrays

```python
# Calculando a média
data = np.array([[1,2],[3,4]])
data.mean()
```

### Desvio padrão em arrays

```python
data.std()
```

### Identificando valores

```python
print(data.min()) # Identificando o valor mínimo
print()

print(data.max()) # Identificando o valor máximo
print()

print(data.sum()) # Somando todos os valores da array numpy
print()

print(np.unique(data)) # Identificando os valores únicos
```
